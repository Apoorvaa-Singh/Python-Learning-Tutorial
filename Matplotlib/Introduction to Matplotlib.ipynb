# Introduction to Matplotlib

matplotlib is an open source Python library built for data visualisation that lets you produce a wide variety of highly customizable charts and graphs;<br>
Matplotlib can plot many data types including base Python sequences,  Numpy Array and pandas series & dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### plot() function

We will be using pyplot module exclusively for plotting. <br>
The plot function creates a line chart by default using index as the x-axis and the list element as the y axis (y values)

```python
# passing a Python list
plt.plot([0,1,4,9,16,25]);
```

When we pass the entire dataframe to the plot function, all the numeric columns are plotted against the index as a separate line. INdex is used as x axis by default

```python
# passing the entire data frame
data={'x':[1,2,3,4,5],'y':[1**2, 2**2,3**2, 4**2, 5**2 ], 'z':[1**3, 2**3,3**3, 4**3, 5**3 ]}
df=pd.DataFrame(data)
```

```python
plt.plot(df);
```

### PLotting framework

Matplotlib has two plotting methods: <br>
1. Pyplot API: Charts are created with plot function and modified with additional function. It is greate for builting quick prototype, mainly for exploration, but not representatiton.
> * Works directly by calling plot() function as above <br>
<br>
2. Objected oriented: In this method the charts are created by defining a plot object,and modified using figure and axis methods. More control of whats happening with our figure. Highly customizable. <br>
> * Creating a figure object and assigning it to a variable : fig = plt.figure()<br>
> * Add chart/axes object to the figure object: ax = fig.add_sublpot()<br>
> * call axes plot method to draw the chart<br>

Subplots(): Lets you create figure and axes in the same line of code. [Better way of doing it] <br>
you can then use figure ans axis methods to customise different elements in the plot. <br>

We built these chart layer by layer. 

```python
fig = plt.figure()
ax = fig.add_subplot();
```

By default the chart is going to be 1 by 1 sqaure without any data plotted. You can then use the plot function on ax object to get the plotted chart <br>
One of the Major difference in the object Oriented plotting is that it use the setter/ getter method. 

```python
# PLotting with objected oriented approach

fig, ax = plt.subplots() # fig is the blank canvas
ax.plot(df['z']); # ax is the specific chart we are plotting

fig.suptitle("Figure Level Title");
ax.set_title("Chart Level Title");
```
