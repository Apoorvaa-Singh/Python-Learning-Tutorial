# Introduction to NumPy and Pandas

Both these libraries are two critical Python libraries that help structure data in arrays & DataFrame and contains a lot of built-in functions for the data analysis. <br>
Pandas is Python's most widely used library for data analysis and contains functions for accessing, aggregating, joining and analyzing data. Its data structur, DataFrame is analogous to SQL tables/ Excel Worksheet.<br>
NumPy is an open source library that is the universal standard for working with the numerical data in Python and forms the foundation of the other libraries like - Pandas and Scikit-learn. Pandas DataFrames are built on the NumPy array and can leverage NumPy functions.    

## NumPy Arrays

NumPy arrays are fixed-size containers of items that are more efficient than Python lists or tuple for data processing
1. They only store a single data types. 
2. They can be one dimensional or multi-dimensional
3. Array elements can be modified but the size of the array can't change. 

When a NumPy array is created, they are stored in the contiguos memmory location
<br>
NumPy arrays are a ndarray Python datatype, which stands for n-dimensional arrays.  

### Array Properties 

NumPy arrays have the following key properties:



> 1. **ndim** - the number of dimensions/axes in the array
2. **shape** - the size of the array for each dimension
3. **size** - the total number of elements in the array
4. **dtype** - datatype of the elements in the array. 




### Creating a NumPy arrays

We can use functions ``` np.array()``` to convert Python list and tuples into the NumPy array.<br>
we can create a multi dimensional array with a nested list, where each nested list is a dimension. for this case the nested list should be of same length.  

ones - creates an array of ones of given size, as a float by default. <br>
``` np.ones((rows, cols), dtype)```


zeros - creates an array of zeros of given size, as a float by default. <br>
``` np.zeros((rows, cols), dtype)```

arange - creates an array of integers with given start and stop values and a step size. Only stop is mandatory argument and is not inclusive. 
``` np.arange(start, stop, step)```

linspace - creates an array of floats with given start and stop values with n elements, separated by consistent step size. Stop is inclusive. 
``` np.linspace(start, stop, n)```

### Methods on Array

**Transponse** ``` array.T``` helps in transposing the dataset, where the rows are converted into columns and vice verse. 

reshape - changes an array into the specified dimesion, if compatible. 
``` array.reshape(rows, cols)```
